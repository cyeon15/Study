```python
import pandas as pd

dataurl = 'https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv'
df = pd.read_csv(dataurl, index_col=0)
df.head()
```

**Question 1**

인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라. (날짜기준, 중복포함)

```python
Ans = list(df.loc[df.channelId.isin(df.channelId.value_counts().head(10).index)].channelTitle.unique())
Ans
```

**Question 2**

논란으로 인기동영상이 된 케이스를 확인하고 싶다. dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하라.

```python
Ans = list(df[df['dislikes'] > df['likes']].channelTitle.unique())
# Ans = list(df.loc[df.likes < df.dislikes.channelTitle.unique()])
Ans
```
