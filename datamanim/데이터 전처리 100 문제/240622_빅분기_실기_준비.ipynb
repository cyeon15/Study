```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/Jeju.csv'

# 데이터 로드
df = pd.read_csv(url, encoding='euc-kr')
df.tail(3)
```

```python
df.dtypes
# 수치형 변수를 가진 컬럼을 출력
df.select_dtypes(exclude=object).columns
```

```python
# 범주형 변수를 가진 컬럼을 출력
df.select_dtypes(include=object).columns
```

```python
# 결측치 파악
df.isna().sum()
```

```python
# dataframe의 전반적인 정보 확인
df.info()
```

```python
# dataframe 분포 확인
df.describe()
```

```python
# 특정 컬럼값 확인
df.거주인구
df['거주인구']
```

```python
# 4분위 범위(IQR) 계산 함수
# IQR = Q3(3사분면) - Q1(1사분면) = 최대값 - 최소값
df['평균 속도'].quantile(q=0.75) - df['평균 속도'].quantile(q=0.25)
```

```python
# 컬럼의 유일값 개수 출력
df['읍면동명'].nunique()
# df.nunique()
```
